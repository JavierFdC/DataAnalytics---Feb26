#### # El final de una linea termina un stament

``` python
midpoint = 5
```
Esto linea es una operación de asignación. Hemos creado una variable llamada ``midpoint`` y le hemos asignado un valor de ``5``.

Cuando se termina la linea se termina el *statement* de python. Es decir el fin de linea es la forma de acabar un stament o comando en python. 

Esto lo diferencia de otros lenguajes como C,C++, javascript donde tenemos que terminar cada statement con algun simbolo definido como un punto y coma(``;``).

Para continuar una comando o statement en una linea nueva en Python podemos utlizar la barra  invertida "``\``" :

````python
x = 1 + 2 + 3 + 4 +\
    5 + 6 + 7 + 8
print(x)
````

También se puede continuar una expresión en una línea nueva, cuando estamos en expresiones entre paréntesis:

````python
x = (1 + 2 + 3 + 4 +
     5 + 6 + 7 + 8)

x
````

La mayoria de guías de estilo recomiendan la segunda versión para escribir expresiones o statements multilinea.

#### El punto y aparte puede ser utilizado para terminar una instrucción

#### Identación: El espacio en blanco importa! ```Y MUCHO```!!

A continuación en nuestro script tenemos el siguiente bloque de código:

``` Python
for i in range(10):
    if i < midpoint:
        lower.append(i)
    else:
        upper.append(i)
```
Aquí estamos utilizando en el código herramientas de control de flujo que veremos más adelante. Básicamente estamos utilizando un loop, que sirve para ejecutar repetitivamente el mismo código, y expresiones de condición. De nuevo no hace falta que entiendas la semántica, pero a nivel de sintaxis es un buen ejemplo para entender como funciona la identación. Por ahora consideralo como una demostración de como funciona la identación. Como ves algunas lineas después del for empiezan unos espacios después, y lo mismo ocurre con las lineas "dentro" del if y el else. Eso indica que existen diferentes bloques de código.

En programación, un *bloque* de código es una serie de intrucciones que deben tratarse o ejecutarse como una unidad.

En Python los bloques de código se definen por su *identación*, es decir el espacio que existe desde el comienzo de linea, hasta el comienzo del código:

``` python
for i in range(100):
    # La identación indica que estamos ante un nuevo bloque de código
    total += i
```
En Python, los bloques de código identados van precedidos siempre por dos puntos en la line anterior(``:``) .

La identación facilita la lectura del código, y lo simplifica, algo que muchas personas encuentran útil en Python. Pero a veces puede traer problemas para las personas que se inician en el lenguaje. Por ejemplo los dos códigos que siguen generan resultados diferentes:

```python
>>> if x < 4:         >>> if x < 4:
...     x = x * 2     ...     x = x * 2
...     print(x)      ... print(x)
```
En el ejemplo de la izquierda , ``print(x)`` está dentro del bloque identado, y sólo se ejecutará si is in the indented block, and will be ``x`` es menor que ``4``.
En el ejemplo de la derecha ``print(x)`` está fuera del bloque identado, y se ejecutará independientemente del valor de ``x``! Pruebalo!

````python
x = 6

if x < 4:
  x = x+2
  print(x)
````

````python
x =6
if x < 4:
  x = x+2
  
print(x)
````

El uso del espacio en blanco en Python es algo que a veces causa cierta sorpresa a personas que viene  de otros lenguajes, pero se puede decir que lleva a código más consistente y legible, porque hace que los diferentes bloques sean identificables de una manera muy visual.

Por último remarcar una cuestión importante: la **cantidad** de espacio en blanco es decisión del programador, pero ha de ser consistente durante todo el script/programa.

#### Los espacios  *Dentro* de las lineas  *No Importan*
Si bien los espacios son importantes *antes* de las lineas (indicando bloques de código), los espacios en blanco *dentro* de las lineas, o en  las expresiones no influyen en el funcionamiento.
Por ejemplo las tres formas de escribir el siguiente código dan el mismo resultado:

````python
x=1+2
x = 1 + 2
x             =        1    +                2
````

Abusar de esto sin embargo puede llevarnos a tener problemas de legibilidad, y de hecho es una de las principales maneras que se utilizan para ofuscar código.
Usar espacios también puede llevarnos a tener código m´s legible, sobre todo cuando encontramos operaciones con varios operadores o signos matemáticos. Por ejemplo en las dos expresiones siguientes estamos haciendo una exponencial con un número negativo:

``` python
x=10**-2
```
to
``` python
x = 10 ** -2
```
La segunda expresión es bastante más legible. La mayoria de guías de estilo recomiendan utilizar un espacio sencillo alrededor de signos de operacación binarios, y ningun espacio alrededor de operadores unarios.
