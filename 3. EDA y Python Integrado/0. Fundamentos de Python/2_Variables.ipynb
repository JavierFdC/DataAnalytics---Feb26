# Semántica Básica en Python: Variables y Objetos

Esta sección empezará a cubrir la semántica básica del lenguaje Python .

En esta veremos la semántica de las *variables* y *objetos*, que son las manera en la que almacenamos, referenciamos y operamos con datos en Python.

## Las variables en Python:

Asignar variables en Python es tan sencillo como seleccionar un nombre para la variable, y utilizar el signo de (``=``) que actúa como operador de asignación en Python y escribir un valor. Por ejemplo:

```python
# Asignar el número 4 a la variable x
x = 4
```


Estamos definiendo un *puntero* llamado ``x`` que apunta a un lugar en la memoria que contiene el valor ``4``.
Esto se debe a que Python es un lenguaje *dinámicamente tipado* y las variables en Python en realidad apuntan a objetos. No hay necesidad de "declarar" la variable o incluso de que la variable apunte a información del mismo tipo. Python la guardará en memoria.


````python
# Las variables no necesitan añadirle el tipo
x = 1
print(x) 
````

````python
type(x)
````

````python
# Una variable puede cambiar de tipo
x = "Hola"
print(x)
````

````python
# Para saber qué tipo de dato es la variable utilizamos la función type()
type(x)
````

````python
# En una misma linea puedes incluir varias variables
x, y = 7, 8
print(x, y)
print(x)
````

## Tipos de datos BÁSICOS en Python:

Asignaremos a las variables distintos valores para definir el tipo de dato

````python
# Tipo de dato numérico entero (int)       
edad = 30
numero_manzanas= 12

print(edad + numero_manzanas)
````

````python
type(edad)
````

````python
# Tipo de dato numérico Real (float)   
altura = 1.75
peso = 65.5

imc = peso / (altura*altura) # calculamos el índice de masa corporal

print(imc)
````

````python
type(altura)
````

````python
# Tipo de dato alfanumérico (str)   
texto = "Esto es un texto"

print(texto)
````

````python
type(texto)
````

````python
# Los datos tipo texto se pueden escribir de varias formas:
a = "Texto en una linea"
print(a)

b = """
Texto
en 
varias
lineas
"""
print(b)

c = '''
Otro texto en varias
lineas
'''
print(c)
````

````python
# Tipo de dato booleano (bool)
x= 5>3      # Esto es true porque 5 es mayor que 3
y = 2==4    # Esto es false porque 2 es distinto de 4
print(x)
print(y)
````

````python
type(x)
````

¿Podemos forzar que un dato como 3 lo interprete como texto?
La respuesta es si, aplicando la siguiente función del ejemplo

````python
# Mostrar que la variable a con valor 3 sea de tipo texto
a = str(3)    
````

````python
print(a, type(a))
````

También que un valor lo identifique como entero o como real

````python
# entero
b = int(3)    
print(b, type(b))
````

````python
# real
c = float(3)  
print(c, type(c))
````

Modificar valores en una variable

````python
c = 'Hola'      # ahora x es un string o cadena de caracteres
print(type(c))
print(c)
````

Pero también hay una consecuencia de esta aproximación a las "variables como punteros" y es algo de lo que tenemos que ser concientes.
Si tienes dos variables que apuntan al mismo objeto *mutable*, si cambias uno de los objetos, también cambiaréis el otro objeto!
Por ejemplo si creamos una lísta y la modificamos, vamos a ver que ocurre:

````python
# Otro tipo de ejemplo. Una variable que tiene como valor otra variable
x = 20
y = x
print(y)
````
